```python
plt.figure(figsize=(10,5))
sns.countplot(y='brand', data=data, order=data['brand'].value_counts().index)
plt.title("Brand Distribution")
plt.show()
```

```python
sns.countplot(x='os', data=data)
plt.title("Operating System Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(x='warranty', data=data)
plt.title("Warranty Distribution")
plt.show()
```

```python
sns.histplot(data['price'], kde=True)
plt.title("Price Distribution")
plt.show()
```

```python
sns.histplot(data['ratings'], bins=10, kde=True)
plt.title("Ratings Distribution")
plt.show()
```

```python
sns.scatterplot(x='price', y='specs_score', data=data)
plt.title("Price vs Specs Score")
plt.show()
```
